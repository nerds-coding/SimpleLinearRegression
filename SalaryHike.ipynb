```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
sns.lmplot(x='YearsExperience',y='Salary',data=sal)
plt.show()
```

```python
x=sns.PairGrid(sal,height=5)
x.map(plt.scatter,s=90)

plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
x = np.array([sal.YearsExperience]).reshape(-1,1)
y = np.array([sal.Salary]).reshape(-1,1)
```

```python
xTrain,xTest,yTrain,yTest = train_test_split(x,y,test_size=0.2)
```

```python
reg = LinearRegression()
```

```python
ft = reg.fit(xTrain,yTrain)
```

```python
print(ft.intercept_)
print(ft.coef_)
```

```python
yPred = ft.predict(xTest)
```

```python
pred = pd.DataFrame({'Actual':yTest.flatten(),'Predicted':yPred.flatten()})
```

```python
plt.figure(figsize=(18,9))

sns.lineplot(data=pred)

plt.show()
```

```python
from sklearn.metrics import mean_squared_error, r2_score
```

```python
mse = mean_squared_error(yTest,yPred)
r2 = r2_score(yTest,yPred)
```

```python
print(mse)
print(r2)
```
