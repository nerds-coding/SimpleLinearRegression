```python
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib import style
import numpy as np
```

```python
sns.pairplot(data=cal,height=5, aspect=1)

plt.show()
```

```python
sns.lmplot(data=cal,x='Calories Consumed',y='Weight gained (grams)')
plt.show()
```

```python
from sklearn import preprocessing,svm
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
x = np.array([cal['Calories Consumed']]).reshape(-1,1)
y = np.array([cal['Weight gained (grams)']]).reshape(-1,1)
```

```python
xTrain,xTest,yTrain,yTest = train_test_split(x,y,test_size=0.2)
```

```python
reg = LinearRegression()
```

```python
lr = reg.fit(xTrain,yTrain)
```

```python
print(lr.intercept_)
print(lr.coef_)
```

```python
yPred = lr.predict(xTest)
```

```python
data = pd.DataFrame({'Actual':yTest.flatten(),'Predict':yPred.flatten()})
```

```python
sns.lineplot(data=data)
plt.show()
```

```python
from sklearn.metrics import mean_squared_error,r2_score
```

```python
mse = mean_squared_error(yTest,yPred)
r2 = r2_score(yTest,yPred)
```
